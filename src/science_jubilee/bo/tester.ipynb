```python
import paramiko
import cv2
import numpy as np
import os 

def capture_image_via_ssh(ip, username, password, save_dir):
    try:
        # Set up SSH client
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(ip, username=username, password=password)

        # Run the command to capture the image using fswebcam
        stdin, stdout, stderr = ssh.exec_command('fswebcam -r 640x480 --jpeg 85 /tmp/image.jpg')

        # Wait for the command to complete
        stdout.channel.recv_exit_status()

        # Fetch the image file from Raspberry Pi to Windows
        sftp = ssh.open_sftp()
        
        if save_dir:
            os.makedirs(save_dir, exist_ok=True)  # Create the directory if it doesn't exist
            save_path = os.path.join(save_dir, 'captured_image.jpg')
        else:
            save_path = 'captured_image.jpg'  # Default to the current directory

        sftp.get('/tmp/image.jpg', save_path)  # Save the file to the specified location
        sftp.close()
        ssh.close()
        
    except paramiko.SSHException as e:
        print(f"Error capturing image via SSH: {e}")
    except FileNotFoundError as e:
        print(f"Error: {e}")
        
        # Read the image using OpenCV
        image = cv2.imread('captured_image.jpg')

        # Display the image using OpenCV
        cv2.imshow("Captured Image", image)
        cv2.waitKey(0)  # Wait for a key press
        cv2.destroyAllWindows()  # Close the image window
        
        print("Image captured and displayed successfully.")
        

# Replace with your Raspberry Pi's IP address, username, and password
capture_image_via_ssh('192.168.1.2', 'axo', 'Sun92023', 'C:/science-jubilee/axo/data')
```

```python
import paramiko
import cv2
import numpy as np
import os

def capture_image_via_ssh(ip, username, password, save_dir):
    try:
        # Set up SSH client
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(ip, username=username, password=password)

        # Create a Python script on the Raspberry Pi to capture an image using OpenCV
        capture_script = """
import cv2
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise Exception("Could not open webcam")
ret, frame = cap.read()
if not ret:
    raise Exception("Failed to capture image")
cv2.imwrite('/tmp/image.jpg', frame)
cap.release()
"""
        # Write the script to a temporary file on the Raspberry Pi
        sftp = ssh.open_sftp()
        with sftp.file('/tmp/capture_image.py', 'w') as f:
            f.write(capture_script)
        sftp.close()

        # Run the Python script on the Raspberry Pi
        stdin, stdout, stderr = ssh.exec_command("python3 /tmp/capture_image.py")
        stdout.channel.recv_exit_status()  # Wait for the command to complete

        # Fetch the captured image from the Raspberry Pi to the local system
        if save_dir:
            os.makedirs(save_dir, exist_ok=True)  # Create the directory if it doesn't exist
            save_path = os.path.join(save_dir, 'captured_image.jpg')
        else:
            save_path = 'captured_image.jpg'  # Default to the current directory

        sftp = ssh.open_sftp()
        sftp.get('/tmp/image.jpg', save_path)
        sftp.close()

        # Cleanup temporary files on the Raspberry Pi
        ssh.exec_command("rm /tmp/capture_image.py /tmp/image.jpg")
        ssh.close()

        # Read the image using OpenCV
        image = cv2.imread(save_path)
        if image is None:
            raise ValueError(f"Failed to load the image from {save_path}")

        # Display the image using OpenCV
        cv2.imshow("Captured Image", image)
        cv2.waitKey(0)  # Wait for a key press
        cv2.destroyAllWindows()  # Close the image window

        print("Image captured, transferred, and displayed successfully.")

    except paramiko.SSHException as e:
        print(f"SSH Error: {e}")
    except FileNotFoundError as e:
        print(f"File Error: {e}")
    except ValueError as e:
        print(f"Image Error: {e}")

# Replace with your Raspberry Pi's IP address, username, and password
capture_image_via_ssh('192.168.1.2', 'axo', 'Sun92023', 'C:/science-jubilee/axo/data')
```

```python
import paramiko
import cv2
import os

def get_next_filename(save_dir, base_name="image"):
    """Generate the next integer-based filename."""
    existing_files = os.listdir(save_dir)
    image_numbers = [
        int(f.split('.')[0].replace(base_name, ""))
        for f in existing_files
        if f.startswith(base_name) and f.endswith(".jpg") and f[len(base_name):].isdigit()
    ]
    next_number = max(image_numbers, default=0) + 1
    return os.path.join(save_dir, f"{base_name}{next_number}.jpg")

def capture_image_with_incrementing_filenames(ip, username, password, save_dir):
    try:
        # Set up SSH client
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(ip, username=username, password=password)

        # Python script to capture an image
        capture_script = """
import cv2
import time
# Open camera
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise Exception("Could not open webcam")
# Set explicit camera parameters to match video streaming quality
cap.set(cv2.CAP_PROP_EXPOSURE, -5)  # Set exposure (if supported by the camera)
#cap.set(cv2.CAP_PROP_GAIN, 0)       # Set gain
cap.set(cv2.CAP_PROP_BRIGHTNESS, 200)  # Adjust brightness
cap.set(cv2.CAP_PROP_CONTRAST, 50)  # Adjust contrast
cap.set(cv2.CAP_PROP_SATURATION, -10)  # Adjust saturation
#cap.set(cv2.CAP_PROP_WHITE_BALANCE_BLUE_U, 4000)  # Set white balance (blue)
#cap.set(cv2.CAP_PROP_WHITE_BALANCE_RED_V, 4500)   # Set white balance (red)
# Let the camera stabilize

# Capture a single frame
ret, frame = cap.read()
if not ret:
    raise Exception("Failed to capture image")
# Save the captured frame
cv2.imwrite('/tmp/image.jpg', frame)
# Release the camera
cap.release()
        """
        # Write the script to the Raspberry Pi
        sftp = ssh.open_sftp()
        with sftp.file('/tmp/capture_image.py', 'w') as f:
            f.write(capture_script)
        sftp.close()

        # Execute the script on the Raspberry Pi
        stdin, stdout, stderr = ssh.exec_command("python3 /tmp/capture_image.py")
        stdout.channel.recv_exit_status()  # Wait for the command to complete

        # Ensure the save directory exists
        os.makedirs(save_dir, exist_ok=True)

        # Determine the next filename
        save_path = get_next_filename(save_dir)

        # Fetch the captured image to the local machine
        sftp = ssh.open_sftp()
        sftp.get('/tmp/image.jpg', save_path)
        sftp.close()

        # Clean up temporary files on the Raspberry Pi
        ssh.exec_command("rm /tmp/capture_image.py /tmp/image.jpg")
        ssh.close()

        print(f"Image saved to {save_path}")

    except paramiko.SSHException as e:
        print(f"SSH Error: {e}")
    except FileNotFoundError as e:
        print(f"File Error: {e}")
    except ValueError as e:
        print(f"Image Error: {e}")

# Example usage
capture_image_with_incrementing_filenames('192.168.1.2', 'axo', 'Sun92023', 'C:/science-jubilee/axo/data')
```

```python
import paramiko
import cv2
import os
def capture_frame_via_ssh(ip, username, password, save_dir, video_url=None):
    """
    Captures a frame from a video stream running on a remote Raspberry Pi and transfers it to the local machine.
    :param ip: IP address of the Raspberry Pi.
    :param username: Username for the Raspberry Pi.
    :param password: Password for the Raspberry Pi.
    :param save_dir: Local directory to save the captured image.
    :param video_url: URL of the video stream on the Raspberry Pi.
    """
    try:
        # Set up SSH client
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(ip, username=username, password=password)
        # Python script for Raspberry Pi to capture a frame from video stream
        capture_script = f"""
import cv2
cap = cv2.VideoCapture("{video_url or 0}")  # Use 0 for local webcam or video stream URL
if not cap.isOpened():
    raise Exception("Could not open video stream.")
# Read a frame from the video stream
ret, frame = cap.read()
if not ret:
    raise Exception("Failed to capture frame from video stream.")
# Save the frame locally on the Raspberry Pi
cv2.imwrite('/tmp/remote_captured_frame.jpg', frame)
cap.release()
"""
        # Write the script to the Raspberry Pi
        sftp = ssh.open_sftp()
        with sftp.file('/tmp/capture_frame.py', 'w') as f:
            f.write(capture_script)
        sftp.close()
        # Execute the script on the Raspberry Pi
        stdin, stdout, stderr = ssh.exec_command("python3 /tmp/capture_frame.py")
        stdout.channel.recv_exit_status()  # Wait for the command to complete
        # Fetch the image to the local machine
        if save_dir:
            os.makedirs(save_dir, exist_ok=True)
            save_path = os.path.join(save_dir, 'remote_captured_frame.jpg')
        else:
            save_path = 'remote_captured_frame.jpg'
        sftp = ssh.open_sftp()
        sftp.get('/tmp/remote_captured_frame.jpg', save_path)
        sftp.close()
        # Clean up the temporary script and image on the Raspberry Pi
        ssh.exec_command("rm /tmp/capture_frame.py /tmp/remote_captured_frame.jpg")
        ssh.close()
        print(f"Frame captured remotely and saved to {save_path}")
        # Optionally display the image
        image = cv2.imread(save_path)
        cv2.imshow("Captured Frame", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except Exception as e:
        print(f"Error: {e}")
# Example Usage for Raspberry Pi
capture_frame_via_ssh(
    ip='192.168.1.2',
    username='axo',
    password='Sun92023',
    save_dir='C:/science-jubilee/axo/data',
    video_url='http://192.168.1.4:5000/video_feed'
)
```

```python
import cv2
import paramiko
import time

def preview_rpi_ssh_webcam(ip, username, password, stream_port=5000):
    """
    Stream Raspberry Pi webcam feed over SSH and preview it on Windows.
    """
    try:
        # Connect to the Raspberry Pi via SSH
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(ip, port=22, username=username, password=password)

        print("Connected to Raspberry Pi via SSH.")

        # Start the webcam stream using ffmpeg
        ffmpeg_command = f"ffmpeg -f v4l2 -i /dev/video0 -vf 'format=yuv420p' -preset ultrafast -f mpegts tcp://0.0.0.0:{stream_port}"
        ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command(ffmpeg_command)

        # Allow the stream to initialize
        time.sleep(2)

        # Access the stream on the Windows machine
        stream_url = f"tcp://{ip}:{stream_port}"
        cap = cv2.VideoCapture(stream_url)

        if not cap.isOpened():
            print("Error: Could not open the stream. Check if ffmpeg is running.")
            ssh.close()
            return

        print("Press 'q' to quit the preview.")

        while True:
            ret, frame = cap.read()
            if not ret:
                print("Error: Could not read frame from stream.")
                break

            cv2.imshow("Raspberry Pi Webcam Stream", frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        # Cleanup
        cap.release()
        cv2.destroyAllWindows()

        # Stop the ffmpeg process
        ssh.exec_command("pkill -f ffmpeg")
        ssh.close()

    except Exception as e:
        print(f"Error: {e}")

# Replace with your Raspberry Pi details
preview_rpi_ssh_webcam('192.168.1.2', username='axo', password='Sun92023')
```

```python
import cv2
import paramiko
import time

def start_stream_on_rpi(ip, username, password, port=5000, resolution="640x480", fps=30):
    """
    Starts an MJPG-streamer service on the Raspberry Pi via SSH.
    """
    try:
        # Connect to Raspberry Pi via SSH
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(ip, username=username, password=password)
        
        print(f"Connected to Raspberry Pi at {ip} via SSH.")

        # Construct the MJPG-streamer command
        mjpg_command = (
            f"mjpg_streamer -i 'input_uvc.so -d /dev/video0 -r {resolution} -f {fps}' "
            f"-o 'output_http.so -w /usr/local/www -p {port}'"
        )

        # Execute the command on Raspberry Pi
        ssh.exec_command(mjpg_command)
        print(f"Started MJPG-streamer on Raspberry Pi (http://{ip}:{port}/?action=stream).")
        return ssh  # Keep the SSH session open to avoid stopping the process

    except Exception as e:
        print(f"Error starting MJPG-streamer on Raspberry Pi: {e}")
        return None


def preview_stream_on_windows(rpi_ip, port=5000):
    """
    Preview the live video stream from the Raspberry Pi on Windows using OpenCV.
    """
    try:
        stream_url = f"http://{rpi_ip}:{port}/?action=stream"
        cap = cv2.VideoCapture(stream_url)

        if not cap.isOpened():
            print(f"Error: Could not open the stream at {stream_url}.")
            return

        print("Press 'q' to quit the preview.")
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Error: Could not read frame from the stream.")
                break

            cv2.imshow("Raspberry Pi Webcam Stream", frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    finally:
        cap.release()
        cv2.destroyAllWindows()


def main():
    """
    Main function to start the stream on Raspberry Pi and preview it on Windows.
    """
    # Raspberry Pi connection details
    rpi_ip = "192.168.1.2"
    username = "axo"
    password = "Sun92023"
    port = 5000  # HTTP streaming port

    # Start the stream on Raspberry Pi
    ssh_session = start_stream_on_rpi(rpi_ip, username, password, port=port)

    if ssh_session:
        # Allow some time for the stream to initialize
        time.sleep(3)

        # Preview the stream on Windows
        preview_stream_on_windows(rpi_ip, port)

        # Close the SSH session
        ssh_session.close()
        print("Stopped MJPG-streamer and closed SSH connection.")

if __name__ == "__main__":
    main()
```
